```python
from pathlib import Path
import sys

project_root = Path.cwd().parent
project_root
```

```python
raw_dir = project_root / "data"
parquet_dir = project_root / "data_parquet"
processed_dir = project_root / "data_processed"
models_dir = project_root / "models"
logs_dir = project_root / "logs"
utils_dir = project_root / "utils"

(raw_dir, parquet_dir, processed_dir, models_dir, logs_dir, utils_dir)
```

```python
createIfMissing = [
    parquet_dir,
    processed_dir,
    models_dir,
    logs_dir,
    utils_dir
]

for f in createIfMissing:
    if not f.exists():
        f.mkdir(parents=True, exist_ok=True)
        print(f"Created: {f}")
```

```python
# Add utils folder so we can import spark_init.py and others
if str(utils_dir) not in sys.path:
    sys.path.append(str(utils_dir))

print("utils/ folder added to Python import path")
```
